```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
savyour=pd.read_excel("Savyour User Experience Survey.xlsx")
```

```python
savyour.head()
```

```python
savyour=savyour.replace(np.NaN,"")
```

```python
savyour.head()
```

```python
text=savyour["First Impression"][0]
```

```python
wordcloud=WordCloud(max_font_size=50,max_words=50,background_color="white").generate(text)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
text=" ".join(rev for rev in savyour["First Impression"])
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
stopwords=["many","operational","well","new","time","seemed","was",""]
wordcloud=WordCloud(max_font_size=75,max_words=25,background_color="white",stopwords=None,colormap="Reds").generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
#wordcloud.to_file("first_impression.png")
```

```python
text2=" ".join(rev for rev in savyour["Like"])
stopwords=["one","The","use","and","It","was","i"]
wordcloud=WordCloud(max_font_size=75,max_words=25,background_color="white",stopwords=None,colormap='Reds').generate(text2)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file("Like.png")
```

```python
text3=" ".join(rev for rev in savyour["Problem"])
stopwords=["Difficulty"]
wordcloud=WordCloud(max_font_size=75,max_words=25,background_color="white",stopwords=stopwords,colormap="hot").generate(text3)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
